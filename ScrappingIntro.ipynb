```python
from scrapy import Selector
```

```python
html = '''
    <html>
        <body>
            <div class="hello python">
                <p>Hello world </p>
            </div>
            <p>Enjoy webscrapping </p>
        </body>
    </html>
'''
```

```python
sel = Selector(text = html)
```

```python
sel.xpath("//p")
```

### Extracting data from the SelectorList using `extract` method

```python
sel.xpath('//p').extract()
```

Extracting the first element from the list

```python
sel.xpath('//p').extract_first()
```

```python
ps = sel.xpath('//p')
second_p = ps[1]
second_p
```

```python
second_p.extract()
```

### HTML text to Selector

```python
import requests
url = 'https://www.sharesansar.com/today-share-price'
html= requests.get(url).content
```

```python
sel = Selector(text = html)
```

```python
sel.xpath('/html/body//div')
```

```python
sel.xpath('/html/body//div').extract()
```

```python
sel.xpath('/html/body//div').extract_first()
```
